# 1. Fizzbuzz

````python
# Write a program that prints numbers 1-100

# Multiples of 3 = Fizz (instead of the number) 3,6,9,12,15 = Fizz
# Multiples of 5 = Buzz (instead of the number) 5,10,15,20,25,30 = Buzz
# Multiples of 3 and 5 = FizzBuzz (instead of the number) 15,30 = FizzBuzz


# While loop


x = 0
while x<100:
    x = x+1
    
    if x % 3 == 0 and x % 5 == 0:
        print('FizzBuzz')
        
    elif x % 3 == 0:
        print('Fizz')
        
    elif x % 5 == 0:
        print('Buzz')
        
    else:
        print(x)
````

# 2. Prime number checker



````python
# Write a Python program that takes in a positive integer 
# and checks if it's a prime number or not
# A prime number (greater than 1) has no positive divisors other than 1 and itself


prime_number_check = (int(input("Enter a positive integer, to check if it's a prime number or not")))


# Use of Modulo, a prime number can only be divided by 1 and itself.
if prime_number_check > 1:
    is_prime = True                                                       # is_prime = 
    
    
    for i in range(2, int(prime_number_check ** 0.5) + 1):                 # This part is a little confusing
        if prime_number_check % i == 0:
            is_prime = False
            break


    if is_prime:
        print(f"{prime_number_check} is a prime number")
    else:
        print(f"{prime_number_check} is NOT a prime number")
        
        
else:
    print(f"{prime_number_check} is NOT a prime number")
````

## Explanation of the code:

This code checks whether a given number, `prime_number_check`, is a prime number or not. Let's go through it step by step:

1. The code begins with an `if` statement: `if prime_number_check > 1:`. 
**This condition checks if `prime_number_check` is greater than 1 because prime numbers are defined as numbers greater than 1 that have no positive divisors other than 1 and themselves.**

<br>

2. If the condition in the `if` statement is true, the code initializes a boolean variable called `is_prime` and sets it to `True`. This variable will be used to keep track of whether the number is prime or not.

<br>

3. Next, a `for` loop is used to iterate over the range from 2 to the square root of `prime_number_check` (inclusive). The square root of `prime_number_check` is calculated using the expression `int(prime_number_check ** 0.5) + 1`.

<br>

4. Inside the `for` loop, the code checks if `prime_number_check` is divisible by the current value of `i`. If the remainder of the division (`prime_number_check % i`) is equal to 0, it means that `prime_number_check` is divisible by `i` and hence not a prime number. In this case, the code sets `is_prime` to `False` and breaks out of the loop using the `break` statement.

<br>

5. After the `for` loop, the code checks the value of `is_prime`. If it is still `True`, it means that no divisors were found for `prime_number_check`, and therefore, it is a prime number. In this case, the code prints the message: `"{prime_number_check} is a prime number"`.

<br>

6. If `is_prime` is `False`, it means that a divisor was found during the iteration, and the number is not a prime number. In this case, the code prints the message: `"{prime_number_check} is NOT a prime number"`.

<br>

7. If the initial condition in the `if` statement (step 1) is not true, meaning `prime_number_check` is less than or equal to 1, the code directly prints the message: `"{prime_number_check} is NOT a prime number"`. This is because numbers less than or equal to 1 are not considered prime.


To summarize, the code takes a number `prime_number_check` as input and determines whether it is a prime number or not by checking if it has any divisors other than 1 and itself.

# 3. Fibonacci Sequence



````python
# Write a Python program to generate the Fibonacci sequence up to n.
# The Fibonacci Sequence is a series of numbers in which each number is the sum of the two preceding ones



# User input for 'n'
n = int(input("Enter a Fibonacci number: "))  

# Initialize the first two numbers of the sequence
a = 0
b = 1

# Print the initial values of 'a' and 'b'
print(a)
print(b)



# Using a Loop, generate subsequent Fibonacci numbers until the sum of 'a' and 'b' exceeds 'n'
while a + b <= n:
    c = a + b               # Calculate the next Fibonacci number
    print(c)                # Print the next number in the sequence
    

# Update 'a' and 'b' for the next iteration of the loop
    a = b          
    b = c 
````

Certainly! To solve the problem of generating the Fibonacci sequence up to a given number 'n' without using any functions in Python, you can use a loop and a few variables to keep track of the sequence. Here's a hint on how you can approach it:

2. Print the initial values of 'a' and 'b' to start the sequence.

3. Use a loop to generate the subsequent Fibonacci numbers. The loop should continue until the sum of 'a' and 'b' exceeds the given number 'n'.

4. Inside the loop, calculate the next Fibonacci number by adding 'a' and 'b'. Assign this sum to a new variable, let's call it 'c'.

5. Print the value of 'c' to display the next number in the sequence.

6. Update the values of 'a' and 'b' by shifting 'b' to 'a' and 'c' to 'b'. This will prepare for the next iteration of the loop.

7. Repeat steps 4-6 until the sum of 'a' and 'b' exceeds 'n'.

By following these steps, you should be able to generate the Fibonacci sequence up to the given number 'n' without using any functions. Good luck with your implementation!

# 4. Factorial calculator

````python
# Write a Python program to calculate the factorial of a number(non-negative integer).
# The function accepts the number as an argument


num = int(input("Enter a non-negative integer: "))
factorial = 1


if num < 0:
    print("The Factorial is undefined for negative numbers.")
    
elif num == 0:
    print("The Factorial of 0 is 1.")
    
else:                                                                 # (1, num + 1): factorial *=1
    for i in range(1, num + 1):
        factorial *= i
    print("The Factorial of", num, "is", factorial)
````

# 5. Palindrome Checker


````python
# Write a Python program to check whether a given string is a palindrome or not
# A palindrome can be read the same forward and backward, ignoring spacing, punctuation and capitalization

# Palindromes to use = kayak, rotator, deed, racecar, noon



palindrome_check = str.lower(input()) or str.upper(input())         # This allows for differences in Upper and lowercase



if palindrome_check == palindrome_check[::-1]:                      # Using step as -1. palindrome_check[::-1] --> variable in reverse
    print(f"{palindrome_check} is a palindrome.")
    
else:
    print(f"{palindrome_check} is NOT a palindrome.")
````
